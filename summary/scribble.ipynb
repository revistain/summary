### 처음보는 파이썬 문법 정리하는 곳

- .map(type, list)
- .map(function) # function에 따라 앞 변수 map

```python
tmp = [1.2, 2.5, 3.1, 4.8]
tmp1 = list(map(int, tmp))
print(tmp1)

# 위 map함수는 아래와 동일하다
for i in range(len(tmp)):
    tmp1[i] = int(tmp[i])
print(tmp1)
```

- .strip(chars) ## default는 공백제거
- .lstrip(chars)
- .rstrip(chars)

```python
str = '   Hi My name is Jungwoo'

print(str.strip('no'))  # 왼쪽 오른쪽 양쪽에서 str에서 모든 'n','o'제거, 만약 차례로 지워가다 'n', 'o'이 아닌 char 만나면 종료
print(str.lstrip())     # 왼쪽만 공백 제거
print(str.rstrip('wo')) # 오른쪽만 'w', 'o' 조합들 제거
```
